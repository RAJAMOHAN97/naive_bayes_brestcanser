```python
data["sex"]=pd.to_numeric(data["sex"], errors='coerce')
data["histologic-type"]=pd.to_numeric(data["histologic-type"], errors='coerce')
data["degree-of-diffe"]=pd.to_numeric(data["degree-of-diffe"], errors='coerce')
data["skin"]=pd.to_numeric(data["skin"], errors='coerce')
data["axillar"]=pd.to_numeric(data["axillar"], errors='coerce')
```

```python
X=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.80, random_state=0)
gnb = GaussianNB()
y_pred = gnb.fit(X_train, y_train).predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, recall_score,confusion_matrix, roc_curve, roc_auc_score
accuracy_score(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```

```python
conf_mat=confusion_matrix(y_test,y_pred)
conf_mat
```

```python
true_positive = conf_mat[0][0]
false_positive = conf_mat[0][1]
false_negative = conf_mat[1][0]
true_negative = conf_mat[1][1]
```

```python
#the formula for Accuracy
Accuracy = (true_positive + true_negative) / (true_positive +false_positive + false_negative + true_negative)
Accuracy
```

```python
# Precison
Precision = true_positive/(true_positive+false_positive)
Precision
```

```python
# Recall
Recall = true_positive/(true_positive+false_negative)
Recall
```
